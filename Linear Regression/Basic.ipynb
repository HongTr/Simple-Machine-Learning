```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Creat 1000 random values (x, y) with noise

```python
X = np.random.rand(1000)
y = 3.6 + 2.1 * X + .25 * np.random.randn(1000) 
plt.scatter (X, y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

Scikit - learn

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X.reshape(-1, 1), y.reshape(-1,1))

w0 = model.intercept_[0]
w1 = model.coef_[0][0]

print("y = {} + {}x".format(w0, w1))
plt.scatter(X, y)
plt.plot(X, w0 + w1 * X, 'r')
plt.show()
```

Gradient Descent (Batch)

```python
w = np.array([3., 2.]).reshape(1, -1) #Randomly initialize bias
Xbar = np.vstack((np.ones((1000)) , X)).T
ybar = y.reshape(-1, 1)
```

```python
# Batch gradient descent
Epoch = 1000
lr = 0.01 #learning rate
Loss = [] #for debugging
for _ in range(Epoch):
    y_hat = np.dot(Xbar, w.T)
    loss = np.mean(.5 * (y_hat - ybar) * (y_hat - ybar)) #MSE
    Loss.append(loss)
    gradients = np.multiply(Xbar, np.hstack((y_hat - ybar, y_hat - ybar)))
    gradients = np.mean(gradients, axis = 0)
    w -= lr * gradients.reshape(1, -1)
```

```python
plt.plot(Loss)
plt.xlabel('Epoch')
plt.ylabel('Loss values')
plt.show()
```

```python
print("y = {} + {}x".format(w[0][0], w[0][1]))
plt.scatter(X, y)
plt.plot(X, w[0][0] + w[0][1] * X, 'r')
plt.show()
```
